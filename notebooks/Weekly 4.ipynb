```python
def in_range(number):
    return 0 <= number <= 100

value = int(input("Enter an integer: "))
print(in_range(value))
```

```python
def count_case(text):
    upper = 0
    lower = 0

    for char in text:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1

    return upper, lower

message = input("Enter a string: ")
upper, lower = count_case(message)
print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
def remove_last_char(text):
    if len(text) <= 1:
        return text
    return text[:-1]

word = input("Enter a string: ")
print(remove_last_char(word))
```

```python
import statistics

temps = []

for i in range(6):
    temp = input("Enter temperature: ")
    value = float(temp[:-1])
    temps.append(value)

print("Max:", max(temps))
print("Min:", min(temps))
print("Mean:", statistics.mean(temps))
```

```python
import statistics

temps = []

while True:
    temp = input("Enter temperature: ")
    if temp == "":
        break
    value = float(temp[:-1])
    temps.append(value)

if temps:
    print("Max:", max(temps))
    print("Min:", min(temps))
    print("Mean:", statistics.mean(temps))
else:
    print("No temperatures entered.")
```

```python
def c_to_f(c):
    return (c * 9 / 5) + 32

def f_to_c(f):
    return (f - 32) * 5 / 9

# Test
print(c_to_f(0))
print(f_to_c(32))
```
