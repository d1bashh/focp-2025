```python
password1 = input("Enter a new password: ")
password2 = input("Confirm your password: ")

if password1 != password2:
    print("Error: Passwords do not match")
elif len(password1) < 8 or len(password1) > 12:
    print("Error: Password must be between 8 and 12 characters")
else:
    print("Password Set")
```

```python
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

password1 = input("Enter a new password: ")
password2 = input("Confirm your password: ")

if password1 != password2:
    print("Error: Passwords do not match")
elif len(password1) < 8 or len(password1) > 12:
    print("Error: Password must be between 8 and 12 characters")
elif password1.lower() in BAD_PASSWORDS:
    print("Error: Password is too common")
else:
    print("Password Set")
```

```python
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

while True:
    password1 = input("Enter a new password: ")
    password2 = input("Confirm your password: ")

    if password1 != password2:
        print("Passwords do not match")
    elif len(password1) < 8 or len(password1) > 12:
        print("Password must be between 8 and 12 characters")
    elif password1.lower() in BAD_PASSWORDS:
        print("Password is too common")
    else:
        print("Password Set")
        break
```

```python
print("Seven Times Table")
for i in range(13):
    print(f"{i} x 7 = {i * 7}")
```

```python
print ("Times Table")
number = int(input("Enter a number between 0 and 12: "))

if 0 <= number <= 12:
    for i in range(13):
        print(f"{i} x {number} = {i * number}")
else:
    print("Number must be between 0 and 12")
```

```python
print("Times Table")
number = int(input("Enter a number: "))

if abs(number) <= 12:
    if number >= 0:
        for i in range(13):
            print(f"{i} x {number} = {i * number}")
    else:
        for i in range(12, -1, -1):
            print(f"{i} x {number} = {i * number}")
else:
    print("Number must be between -12 and 12")
```
