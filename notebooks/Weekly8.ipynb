```python
def nl(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            for i, line in enumerate(f, start=1):
                print(f"{i}\t{line.rstrip()}")
    except FileNotFoundError:
        print("File not found.")
```

```python
def diff(file1, file2):
    try:
        with open(file1, "r", encoding="utf-8") as f1, open(file2, "r", encoding="utf-8") as f2:
            if f1.read() == f2.read():
                print("The files are the same.")
            else:
                print("The files are different.")
    except FileNotFoundError:
        print("One or both files not found.")
```

```python
def grep(search_string, filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            for line in f:
                if search_string in line:
                    print(line.rstrip())
    except FileNotFoundError:
        print("File not found.")
```

```python
def wc(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            text = f.read()
            lines = text.count("\n")
            characters = len(text)
            print("Lines:", lines)
            print("Characters:", characters)
    except FileNotFoundError:
        print("File not found.")
```

```python
import string

def spell(filename, dictionary_file):
    try:
        with open(dictionary_file, "r", encoding="utf-8") as d:
            dictionary = set(word.strip().lower() for word in d)

        with open(filename, "r", encoding="utf-8") as f:
            text = f.read().lower()

        for p in string.punctuation:
            text = text.replace(p, "")

        words = text.split()

        misspelled = sorted(set(word for word in words if word not in dictionary))

        if misspelled:
            print("Misspelled words:")
            for word in misspelled:
                print(word)
        else:
            print("No spelling errors found.")

    except FileNotFoundError:
        print("File or dictionary not found.")
```
