```python
def to_binary(n):
    return bin(n)[2:]

number = int(input("Enter a positive integer: "))
result = to_binary(number)
print("Binary:", result)
```

```python
def factors(n):
    result = []
    for i in range(1, n + 1):
        if n % i == 0:
            result.append(i)
    return result

number = int(input("Enter an integer: "))
print("Factors:", factors(number))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

number = int(input("Enter an integer: "))

if is_prime(number):
    print(number, "is a prime number")
else:
    print(number, "is not a prime number")
```

```python
def simple_encrypt(message):
    return message.replace(" ", "")[::-1]

message = input("Enter a message: ")
print("Encrypted message:", simple_encrypt(message))
```

```python
import random
import string

def hide_message(message):
    interval = random.randint(2, 20)
    encrypted = []

    for char in message.replace(" ", ""):
        encrypted.append(char)
        for _ in range(interval - 1):
            encrypted.append(random.choice(string.ascii_lowercase))

    return "".join(encrypted), interval

message = input("Enter a message to hide: ")
encrypted, interval = hide_message(message)

print("Encrypted message:", encrypted)
print("Interval used:", interval)
```

```python
def reveal_message(encrypted, interval):
    return encrypted[::interval]

encrypted = input("Enter the encrypted message: ")
interval = int(input("Enter the interval: "))

print("Decrypted message:", reveal_message(encrypted, interval))
```
