```python
import sys

def main():
    args = sys.argv[1:]

    if not args:
        print("Error: No arguments provided.")
        return

    args.sort(key=len)
    print(f"Shortest argument: {args[0]}")

if __name__ == "__main__":
    main()
```

```python
from urllib.request import urlopen, Request
from urllib.error import URLError, HTTPError

def check_url(url: str, timeout: int = 10):
    if not url or not isinstance(url, str):
        return (False, None, "Error: No valid URL provided.")

    # Some websites block requests without a User-Agent
    req = Request(url, headers={"User-Agent": "Mozilla/5.0"})

    try:
        with urlopen(req, timeout=timeout) as resp:
            return (True, resp.status, "Reachable")
    except HTTPError as e:
        return (True, e.code, f"Responded with HTTP error {e.code}")
    except URLError as e:
        return (False, None, f"Not reachable: {e.reason}")
    except Exception as e:
        return (False, None, f"Unexpected error: {e}")

# Example:
check_url("https://www.google.com")

```

```python
def temp_stats(arg_list):

    if not arg_list:
        return "Error: No temperature values provided."

    try:
        temps = [float(x) for x in arg_list]
    except ValueError:
        return "Error: All temperature values must be numbers."

    return (min(temps), max(temps), sum(temps) / len(temps))

temp_stats(["12.5", "15", "9", "20"])
```

```python
import shutil
import os

def backup_file(filename):

    if not filename or not isinstance(filename, str):
        return "Error: No valid filename provided."

    if not os.path.exists(filename):
        return f"Error: File '{filename}' does not exist."

    backup_name = "backup" + filename

    try:
        shutil.copy(filename, backup_name)
        return f"Backup created successfully: {backup_name}"
    except Exception as e:
        return f"Error creating backup: {e}"

with open("example.txt", "w") as f:
    f.write("This is a test file.")

backup_file("example.txt")

```
